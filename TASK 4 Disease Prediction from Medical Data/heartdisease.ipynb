## Title - Heart Disease Detection Using Logistic Regression

### Description
#### The Heart Disease Prediction project uses Logistic Regression machine learning techniques to predict the likelihood of a patient having heart disease based on clinical and medical attributes. The model analyzes features such as age, sex, chest pain type, resting blood pressure, cholesterol level, fasting blood sugar, ECG results, maximum heart rate, exercise-induced angina, and other cardiovascular indicators. Multiple classification algorithms are trained and evaluated to identify the most accurate model for early heart disease detection, enabling healthcare professionals to make informed clinical decisions and improve patient outcomes.

### Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

### Downloading Dataset

```python

heart_data = pd.read_csv('Datasets/heart.csv')   
heart_data.head()
```

## 1 --> Defective Heart

## 0 --> Healthy Heart

```python


X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python

print(X)
```

```python
print(Y)
```

### Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Model Training

### Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
```

```python


model.fit(X_train, Y_train)
print("Model trained successfully!")
```


### Model Evaluation

### Accuracy Score

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(" Accuracy score of the training data is ",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score of the test data is",test_data_accuracy)
```

### Building a Predictive System

```python
input_data = (60,1,0,117,230,1,1,160,1,1.4,2,2,3
)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
     
```

### Rechecking The Trained Model

```python
input_data = (44,0,2,118,242,0,1,149,0,0.3,1,1,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
     
```
